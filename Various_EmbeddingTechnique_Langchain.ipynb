#1) BY OPENAIEMBEDDING

```python
from langchain_openai import OpenAIEmbeddings
encoder=OpenAIEmbeddings(model='text-embedding-3-small')
```

```python
!wget -O 'layoutparser_paper.pdf' 'http://arxiv.org/pdf/2103.15348.pdf'
```

```python
!pip install pymupdf
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
loader=PyMuPDFLoader(file_path='layoutparser_paper.pdf')
```

```python
pages=loader.load()
```

```python
len(pages)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(separators=["\n\n","\n"," ",""],
                                             chunk_size=500,
                                             chunk_overlap=75)
```

```python
chunks=[text_splitter.split_text(page.page_content) for page in pages]
```

```python
texts=[doc for chunk in chunks for doc in chunk ]
```

```python
embeddings=await encoder.aembed_documents(texts=texts)
```

```python
len(embeddings)
```

```python
for vector,text in zip(embeddings,texts):
  print(text," :- ",vector)
```

```python
user_querry=input("Enter your querry :-")
```

```python
querry=await encoder.aembed_query(user_querry)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
result=cosine_similarity([querry],embeddings)
```

```python
print(result)
```

```python
import numpy as np
```

```python
index=np.argmax(result)
```

```python
print(index)
```

```python
print("user querry :- ",user_querry," \n solution :- ",texts[11])
```

```python
min_index=np.argmin(result)
```

```python
print(texts[min_index])
```

#2) HuggingFaceEmbedding

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
model_name = "mixedbread-ai/mxbai-embed-large-v1"
encoder=HuggingFaceEmbeddings(model=model_name)
```

```python
embed_doc=encoder.embed_documents(texts=texts)
querry=encoder.embed_query(user_querry)
```

```python
result_huggingface=cosine_similarity([querry],embed_doc)[0]
```

```python
result_huggingface
```

```python
index_result=np.argmax(result_huggingface)
```

```python
print(index_result)
```
